```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# --- STEP 1: LOAD & CLEAN DATA ---
print("Checking for file...")

if os.path.exists('spam.csv'):
    # Load the file
    df = pd.read_csv('spam.csv', encoding='latin-1')
    
    # Clean the file
    df = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], errors='ignore')
    df.rename(columns={'v1': 'Category', 'v2': 'Message'}, inplace=True)
    df['Spam'] = df['Category'].map({'spam': 1, 'ham': 0})
    
    print("‚úÖ Data Loaded and Cleaned successfully!")

    # --- STEP 2: PREPARE DATA (SPLIT & VECTORIZE) ---
    # Now that we know df exists, we can safely proceed
    
    # Separate Text (X) and Label (Y)
    X = df['Message']
    Y = df['Spam']

    # Split the data (80% Training, 20% Testing)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

    # Initialize the Vectorizer (Converts text to numbers)
    feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

    # Fit and Transform Training data
    X_train_features = feature_extraction.fit_transform(X_train)

    # Only Transform Test data
    X_test_features = feature_extraction.transform(X_test)

    # Ensure Y is integer type
    Y_train = Y_train.astype('int')
    Y_test = Y_test.astype('int')

    print("‚úÖ Data split and vectorization complete!")
    print(f"Training Data Shape: {X_train_features.shape}")

else:
    # If file is missing, print a clear error instead of crashing
    print("\n‚ùå ERROR: 'spam.csv' was not found.")
    print("Please make sure the file is in this folder:", os.getcwd())
```

```python
from sklearn.naive_bayes import MultinomialNB

# Initialize the model
model = MultinomialNB()

# Train the model using the training data features
model.fit(X_train_features, Y_train)

print("‚úÖ Model trained successfully!")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# --- PART 1: TRAIN THE MODEL (This was missing) ---
print("Training the model...")
model = MultinomialNB()
model.fit(X_train_features, Y_train) 
print("‚úÖ Model trained successfully!")

# --- PART 2: EVALUATE THE MODEL ---
# Now we can use the model to predict
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)

print(f'Accuracy on Test Data: {accuracy_on_test_data:.4f}')

# Show the Classification Report
print("\nClassification Report:")
print(classification_report(Y_test, prediction_on_test_data))

# Plot Confusion Matrix
cm = confusion_matrix(Y_test, prediction_on_test_data)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```
